## AG2 Beta: What's Changing

The AG2 framework is evolving. Key changes in the beta:

- **MemoryStream** — A new approach to conversation state management. Instead of passing entire message histories, agents get a curated stream of relevant memories. This reduces token usage and improves coherence in long conversations.

- **Event-driven architecture** — Moving from synchronous message passing to an event-driven model. Agents can react to events (tool completion, user input, timer) rather than waiting for their turn in a round-robin.

- **Improved handoffs** — More flexible agent transitions with better state preservation across handoffs.

For details, see the AG2 blog: [AG2 Beta](https://docs.ag2.ai/blog/2026-03-16-AG2-Beta).

## Challenges in Production Agents

Building agents is getting easier. Running them reliably in production is still hard. These are the open challenges:

### 1. Reliability
Hallucination cascades are the biggest risk in multi-agent systems. When Agent A hallucinates, Agent B builds on that hallucination, and Agent C validates it. The system confidently produces nonsense. Mitigation: grounding agents with tools, verification steps, and human-in-the-loop at critical points.

### 2. Cost at Scale
Episode 19 covered the basics of cost tracking. At scale, you need more sophisticated optimization: prompt caching, model routing (cheap models for simple tasks, expensive for hard ones), and aggressive context window management.

### 3. Testing Maturity
The field is still developing evaluation frameworks for agents. Unit testing agent behavior is fundamentally different from testing deterministic code. How do you assert that an agent "made a good decision"?

### 4. Safety and Alignment
Ensuring agents act within intended boundaries. A customer service agent shouldn't promise refunds it can't authorize. A code agent shouldn't delete production databases. Guardrails, sandboxing, and permission systems are essential.

### 5. State Management
Long conversations across sessions remain tricky. How much context do you preserve? How do you resume a complex multi-agent workflow that was interrupted?

### 6. Error Handling and Retries
Multi-agent failure modes are unique and complex:
- **Hallucination cascading** — one agent's error propagates through the system
- **Handoff loops** — agents pass tasks back and forth indefinitely
- **Tool call failures** — external APIs go down mid-conversation

Key patterns to watch (content coming in future workshop updates):
- **Fallback agents** — secondary agents that activate when the primary fails
- **Circuit breakers** — automatic disengagement after repeated failures
- **Graceful degradation** — reduce capability rather than failing completely

## Latest Research Directions

The multi-agent field is moving fast. Areas to watch:

- **Agent failure attribution** — When a multi-agent system produces a wrong answer, which agent was responsible? Tracing and attributing errors through complex agent graphs.

- **Multi-agent evaluation frameworks** — Beyond single-agent benchmarks. How do you evaluate the quality of agent collaboration? Metrics for handoff efficiency, context preservation, and task decomposition quality.

- **Reasoning agents (MCTS, LATS)** — You explored these in Episode 22. The frontier is combining structured reasoning with multi-agent collaboration: teams of reasoning agents.

- **Human-agent collaboration patterns** — Moving beyond "human-in-the-loop" to more nuanced collaboration where humans and agents have complementary roles. Adaptive autonomy: agents that know when to ask for help.

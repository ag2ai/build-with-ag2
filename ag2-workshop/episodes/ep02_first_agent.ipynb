## Step 2: Configure your model

Every agent needs a model to think with. In the Beta API, a **config object** describes which model to use.

We use `OpenAIConfig` with `gpt-4.1-mini` — a strong balance of speed, capability, and cost. `temperature=0` makes responses more consistent, which is handy while learning.

`load_dotenv()` reads your `OPENAI_API_KEY` from the `.env` file in the `ag2-workshop/` folder.

```python
from dotenv import load_dotenv

load_dotenv()

from autogen.beta.config import OpenAIConfig

config = OpenAIConfig(model="gpt-4.1-mini", temperature=0)
```

## Step 3: Create your agent

`Agent` is the building block of every AG2 Beta system. Three things get you started:

- **`name`** — identifies the agent
- **`prompt`** — defines its personality and behavior (the system prompt)
- **`config`** — connects it to a model

That's it.

```python
from autogen.beta import Agent

agent = Agent(
    "assistant",
    prompt="You are a helpful AI assistant. Be concise.",
    config=config,
)
```

## Step 4: Ask it something

The Beta API is **async-first**. You talk to an agent with `await agent.ask(...)`, which returns a **reply** object. The text of the answer is in `reply.body`.

> **Async note:** In a notebook you can `await` directly, as below. In a plain `.py` script, wrap your code in an `async def main()` and call `asyncio.run(main())`.

```python
reply = await agent.ask("What are three benefits of using multiple AI agents instead of one?")
print(reply.body)
```

## Step 5: Continue the conversation

The reply object remembers the conversation. Call `.ask()` **on the reply** to send a follow-up — the agent keeps the earlier context.

```python
follow_up = await reply.ask("Give a concrete one-sentence example of the first benefit.")
print(follow_up.body)
```

## Additional content

**What's in a config?** `OpenAIConfig` holds everything about *how* the model is called — the model name, `temperature`, `max_tokens`, streaming, and more. Agents that share a config share those settings; give an agent its own config when it needs different behavior.

**Why async?** Real systems wait on many things at once — model calls, tool calls, other agents. An async API lets all of that overlap instead of blocking. Every `ask()` is a coroutine, so it always comes with `await`.

**`reply.body` vs `reply.content()`** — `reply.body` is the plain text. When an agent has a structured output schema (Episode 9), you instead call `await reply.content()` to get a typed object.

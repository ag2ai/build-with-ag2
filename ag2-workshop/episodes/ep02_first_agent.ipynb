## Step 2: Configure your LLM

Every agent needs a brain. `LLMConfig` tells your agent which model to use — think of it as plugging in the engine.

Here we're using OpenAI's `gpt-4o-mini`, which hits a sweet spot of speed, capability, and cost.

````python
from dotenv import load_dotenv
load_dotenv()

from autogen import ConversableAgent, LLMConfig

llm_config = LLMConfig(api_type="openai", model="gpt-4o-mini")
````

## Step 3: Create your agent

Now for the fun part. `ConversableAgent` is the building block of every AG2 system. Three parameters are all you need to get started:

- **`name`** — identifies the agent in conversations
- **`system_message`** — defines the agent's personality and behavior
- **`llm_config`** — connects the agent to a language model (set via the `with` context manager)

That's it. Three lines and you have a working agent.

````python
with llm_config:
    agent = ConversableAgent(
        name="assistant",
        system_message="You are a helpful AI assistant. Be concise and friendly.",
        human_input_mode="NEVER",
    )
````

## Step 4: Chat with your agent

Let's see this in action:

````python
result = agent.run(message="What are three benefits of using multiple AI agents instead of one?")
print(result.summary)
````

## Try it yourself

What happens if you change the agent's personality entirely? Swap the `system_message` to make the agent a pirate, a scientist, or a poet, then re-run the cell.

For example:
```python
system_message="You are a pirate. Always respond in pirate speak, matey!"
```

The system message is the single most powerful lever you have over agent behavior. Small changes produce dramatically different outputs.

## What just happened?

Under the hood, four things happened in quick succession:

1. The agent received your message
2. It sent the message to the LLM (along with its system message)
3. The LLM generated a response
4. The agent returned the response to you

This is the simplest possible agent — no tools, no collaborators, just an LLM and a system prompt. But even this minimal setup is a foundation you can build on. Every multi-agent system you'll see in this workshop is made of agents just like this one.

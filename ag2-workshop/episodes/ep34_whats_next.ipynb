# Episode 34: What's Next

You've reached the end of the workshop. This last episode is a map — what you've built, where AG2 is heading, and where to go from here.

## What you've built

Across 34 episodes you went from a single `await agent.ask()` to production agent systems:

| Group | What you learned |
|---|---|
| 1 · Fundamentals | `Agent`, tools, task delegation, the `LiveAgent` voice showcase |
| 2 · Agent Harness | the constructor's slots — middleware, assembly, structured output |
| 3 · Multi-Agent Networking | the Hub, and the consulting / conversation / discussion / workflow adapters |
| 4 · Choosing a Pattern | a decision framework for delegation vs. networking |
| 5 · Applications | real apps — triage, research, knowledge bases, web UIs, MCP |
| 6 · Production | telemetry, observers, security, testing, cost, deployment |
| 7 · Advanced | redundancy, structured reasoning, the event stream, migration |

That's a complete toolkit: you can design, build, harden, and ship an agent system.

## Where the ecosystem is heading

AG2 Beta is moving fast. The themes to watch:

- **A2A everywhere.** Agent-to-agent communication across teams and organizations — your agents calling, and being called by, agents you didn't write. Episode 29 was the first taste.
- **Networks at scale.** The Hub model (Group 3) is built for larger topologies — many autonomous agents, richer transition graphs, remote links across hosts.
- **Richer frontends.** AG-UI (Episode 21) and the CopilotKit ecosystem are making streaming, tool-aware agent UIs a drop-in.
- **Real-time and multimodal.** `LiveAgent` (Episode 5) points at where interaction is going — full-duplex voice, and beyond.
- **Beta → stable.** The `autogen.beta` API is converging. Expect the surface you learned here to harden into the stable API.

## Where to go next

- **Docs** — the AG2 Beta documentation is the reference for every class and pattern in this workshop.
- **Playground** — runnable demos and network wireframes that go deeper than the episodes: `beta_conversation`, `parallel_research`, `working_memory`, the `network_*` demos, and the `network/wireframes/` walkthroughs.
- **The runnable examples** — every code block in this workshop has a standalone script in `_runnable_examples/`. Run `run_all.sh` to execute them end to end.
- **Build something.** The fastest way to consolidate this is a project of your own. Start from the application closest to your goal (Group 5) and extend it.

- **Community** — share what you build and follow the roadmap at [github.com/ag2ai/ag2](https://github.com/ag2ai/ag2).

## Thank you

You started this workshop able to call one agent. You're leaving it able to build a *system* of them — observable, secure, tested, and deployed.

Now go build. The most interesting agents haven't been written yet.

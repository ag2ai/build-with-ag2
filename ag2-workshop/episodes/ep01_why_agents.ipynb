## What is an agent?

You've used ChatGPT. You type a question, you get a response. That's a single LLM call — powerful, but passive. It can't go check your database, browse the web, or loop back to reconsider its own answer.

An **agent** wraps that same LLM in a loop with the ability to act. It can:

- **Send and receive messages** — communicating with users and other agents
- **Use AI models** as a backend for reasoning and generating responses
- **Use tools** to take actions in the real world (call APIs, query databases, run code)
- **Request human input** when it needs guidance or approval

The jump from "chatbot" to "agent" is the jump from *answering* to *doing*. In AG2, the fundamental building block is the **`ConversableAgent`** — a single class that can be configured to play any role in a multi-agent system.

## The Agent Paradigm

Back in 1986, Marvin Minsky proposed the "Society of Mind" — the idea that intelligence emerges not from one brilliant module, but from many simple agents cooperating. Decades later, that's exactly what modern multi-agent systems look like.

Watch for these four properties when you run the code later — every agent you build will exhibit them:

| Property | Description |
|---|---|
| **Autonomy** | Operates without constant human intervention |
| **Reactivity** | Responds to changes in its environment |
| **Proactivity** | Takes initiative to achieve goals |
| **Social Ability** | Communicates and collaborates with other agents |

Agent systems have evolved through several generations, each building on the last:

1. **Rule-based** — hard-coded if/else logic
2. **Model-based** — agents that maintain internal state
3. **Multi-agent** — multiple agents collaborating
4. **AI-driven** — LLMs as the reasoning engine
5. **Composable** — mix-and-match agents, tools, and orchestration patterns (where we are today)

We're at an inflection point. Gartner predicts that 40% of enterprise apps will embed AI agents by the end of 2026, up from less than 5% in 2025. This isn't theoretical — it's happening now.

## Why MULTIPLE agents?

Imagine asking one person to research a topic, write the report, review it for accuracy, and then edit the prose. They *can* do it — but the quality suffers because they're wearing too many hats at once.

Now imagine a team: a researcher, a writer, a fact-checker, and an editor. Each person focuses on what they do best. That's the multi-agent advantage.

Here's why it works:

- **Separation of concerns** — each agent has a clear, focused role
- **Specialization** — agents can have different system prompts, tools, and expertise
- **Testability** — you can test and debug each agent independently
- **Composability** — swap agents in and out without rewriting the whole system

Researchers at Tufts University found that multiple agents running GPT-3.5 actually outperformed a single GPT-4 agent — at lower cost. The structure of collaboration matters as much as the raw model capability.

## Trade-offs to consider

Agents aren't magic dust you sprinkle on every problem. Like any architecture decision, there are trade-offs worth thinking through.

**Agents shine when** the task benefits from iteration, collaboration, or role separation. Think complex research, multi-step analysis, content that needs review, or workflows where different steps require different expertise.

**A direct LLM call is better when** you need a quick answer to a straightforward question, you're calling a single function, or latency is critical. If the first response is good enough, agents add unnecessary complexity.

And cost? It's nuanced. Yes, agents make multiple LLM calls — but Skypoint AI found that their multi-agent system reduced Medicare plan analysis from hours to 5-10 minutes. The extra tokens are a rounding error compared to the time saved.

## What is AG2?

**AG2** is an open-source multi-agent framework and the continuation of Microsoft's AutoGen project. With over half a million monthly installs, it has a large and active community behind it.

The core design centers on a few key ideas:

- **`ConversableAgent`** as the universal building block — one class that can be configured for any role
- Built-in support for **tool use, group chat, nested conversations**, and more
- A growing ecosystem of patterns for orchestration, handoffs, and agent composition

Everything in this workshop builds on `ConversableAgent`. Once you understand it, you understand AG2.

## Let's see it in action

Enough theory — let's watch two agents have a conversation. The code below creates a **student** and a **teacher** and has them talk.

Don't worry about understanding every line yet. Just watch the output and notice how each agent stays in its role.

```python
from dotenv import load_dotenv

load_dotenv()

from autogen import ConversableAgent, LLMConfig

# Connect to OpenAI's gpt-4o-mini model
llm_config = LLMConfig({"model": "gpt-4o-mini"})

# Two agents: a student asks, a teacher answers
student = ConversableAgent(
    "student",
    system_message="You are a curious student. Ask one short question about AI agents.",
    llm_config=llm_config,
)
teacher = ConversableAgent(
    "teacher",
    system_message="You are a helpful teacher. Give a brief, clear answer.",
    llm_config=llm_config,
    human_input_mode="NEVER",
)

# Start the conversation — the student sends a message, the teacher replies
result = student.run(
    teacher,
    message="What makes AI agents different from regular chatbots?",
    max_turns=2,
)
result.process()
print(result.summary)
```

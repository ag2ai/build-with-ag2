## What is an agent?

You've used ChatGPT. You type a question, you get a response. That's a single LLM call — powerful, but passive. It can't go check your database, browse the web, or loop back to reconsider its own answer.

An **agent** wraps that same LLM in a loop with the ability to act. It can:

- **Send and receive messages** — communicating with users and other agents
- **Use a model** as its reasoning backend
- **Use tools** to take actions in the real world (call APIs, query databases, run code)
- **Request human input** when it needs guidance or approval

The jump from "chatbot" to "agent" is the jump from *answering* to *doing*. In the AG2 Beta API, the fundamental building block is the **`Agent`** class — a single class you configure to play any role in a system.

## The agent paradigm

Back in 1986, Marvin Minsky proposed the "Society of Mind" — the idea that intelligence emerges not from one brilliant module, but from many simple agents cooperating. Modern multi-agent systems look exactly like that.

Watch for these four properties — every agent you build exhibits them:

| Property | Description |
|---|---|
| **Autonomy** | Operates without constant human intervention |
| **Reactivity** | Responds to changes in its environment |
| **Proactivity** | Takes initiative to achieve goals |
| **Social ability** | Communicates and collaborates with other agents |

We're at an inflection point: Gartner predicts 40% of enterprise apps will embed AI agents by the end of 2026, up from less than 5% in 2025. This isn't theoretical — it's happening now.

## Why MULTIPLE agents?

Imagine asking one person to research a topic, write the report, fact-check it, and edit the prose. They *can* — but quality suffers because they're wearing too many hats at once.

Now imagine a team: a researcher, a writer, a fact-checker, an editor. Each focuses on what they do best. That's the multi-agent advantage:

- **Separation of concerns** — each agent has a clear, focused role
- **Specialization** — different prompts, tools, and expertise per agent
- **Testability** — test and debug each agent independently
- **Composability** — swap agents in and out without rewriting the system

Researchers at Tufts University found that multiple agents running GPT-3.5 outperformed a single GPT-4 agent — at lower cost. The structure of collaboration matters as much as raw model capability.

## Trade-offs to consider

Agents aren't magic dust you sprinkle on every problem.

**Agents shine when** the task benefits from iteration, collaboration, or role separation — complex research, multi-step analysis, content that needs review, workflows where different steps need different expertise.

**A direct LLM call is better when** you need a quick answer to a straightforward question, you're calling a single function, or latency is critical.

And cost? It's nuanced. Agents make multiple LLM calls — but Skypoint AI found their multi-agent system cut Medicare plan analysis from hours to 5–10 minutes. The extra tokens are a rounding error next to the time saved.

## What is AG2?

**AG2** is an open-source multi-agent framework and the continuation of Microsoft's AutoGen project, with a large and active community.

This workshop teaches the **AG2 Beta API** (`autogen.beta`) — the current, preferred surface. It is a clean, **async-first** redesign built on a few ideas:

- **`Agent`** — the universal building block. You give it a name, a prompt, a model config, and optional capabilities.
- **The harness** — an `Agent` is composable: tools, middleware, observers, assembly policies, knowledge, and structured output all plug into the constructor.
- **The Hub** — for true multi-agent systems, autonomous agents communicate over channels rather than function calls.

You run an agent with one line: `reply = await agent.ask("...")`.

Every agent run flows through the same pipeline. Keep this picture in mind — each group of episodes zooms into one part of it.

## Classic vs Beta — at a glance

If you've seen older AG2 code, here's how the surface changed. This workshop uses **Beta only** — the classic column is just for orientation.

| Task | Classic AG2 | AG2 Beta |
|---|---|---|
| Create an agent | `ConversableAgent(...)` | `Agent(name, prompt=..., config=...)` |
| Configure the model | `LLMConfig({"model": ...})` | `OpenAIConfig(model=...)` |
| Run it | `user.run(agent, message=...)` | `reply = await agent.ask(...)` |
| Read the answer | parse the result object | `reply.body` |
| Add a tool | `register_for_llm` + `register_for_execution` | pass a function to `tools=[...]` |

## Let's see it in action

Enough theory — let's watch two agents weigh in on the same topic. We create an **optimist** and a **skeptic**, then ask each for its view.

Don't worry about every line yet — just notice how each agent stays firmly in its role. That role separation is the whole point.

```python
from dotenv import load_dotenv

load_dotenv()

from autogen.beta import Agent
from autogen.beta.config import OpenAIConfig

# One model config, shared by both agents
config = OpenAIConfig(model="gpt-4.1-mini", temperature=0)

# Two agents, two roles
optimist = Agent(
    "optimist",
    prompt="You are an optimist. In two sentences, give the upside of the topic.",
    config=config,
)
skeptic = Agent(
    "skeptic",
    prompt="You are a skeptic. In two sentences, give a caution about the topic.",
    config=config,
)

topic = "letting AI agents book travel autonomously"

# Ask each agent for its view
up = await optimist.ask(f"Topic: {topic}")
print("Optimist:", up.body)
print()
down = await skeptic.ask(f"Topic: {topic}")
print("Skeptic:", down.body)
```
